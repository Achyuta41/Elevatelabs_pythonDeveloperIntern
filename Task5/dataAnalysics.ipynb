```python
# ===========================================
# üß© Task 5: Data Analysis on CSV Files
# Dataset: education_data.csv
# Objective: Analyze school enrollments by sector, type, and gender
# Tools: Python, Pandas, Matplotlib, Seaborn
# ===========================================

# ‚úÖ Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------------------
# 1Ô∏è‚É£ Load and Explore the Dataset
# -------------------------------------------
df = pd.read_csv('school_enrollment_dataset.csv',encoding='windows-1252')  # Replace with your actual file path

print("üîπ First 5 Rows:")
display(df.head())

print("\nüîπ Shape of Data:", df.shape)
print("\nüîπ Column Names:")
print(df.columns.tolist())

print("\nüîπ Basic Info:")
df.info()

# -------------------------------------------
# 2Ô∏è‚É£ Data Cleaning
# -------------------------------------------
# Clean column names (remove spaces, quotes)
df.columns = df.columns.str.strip().str.replace('"', '').str.replace(' ', '_')

# Replace NaN with 0 for numeric columns
df = df.fillna(0)

# Rename confusing columns (optional)
if 'Grand_Total""' in df.columns:
    df.rename(columns={'Grand_Total""': 'Grand_Total'}, inplace=True)

# Preview after cleaning
print("\n‚úÖ Cleaned DataFrame:")
display(df.head())

# -------------------------------------------
# 3Ô∏è‚É£ Feature Engineering
# -------------------------------------------
# Ensure 'Grand_Total' column exists
if 'Grand_Total' not in df.columns:
    # Try alternate column names
    for col in df.columns:
        if 'Grand_Total' in col:
            df['Grand_Total'] = df[col]
            break

# Convert numeric columns to float
df['Grand_Total'] = pd.to_numeric(df['Grand_Total'], errors='coerce').fillna(0)

# -------------------------------------------
# 4Ô∏è‚É£ Data Analysis
# -------------------------------------------

# üè´ Total Students by School Type
school_type_total = df.groupby('School_Type')['Grand_Total'].sum().sort_values(ascending=False)
print("\nüè´ Total Students by School Type:\n", school_type_total)

# üéì Total Students by Education Sector
sector_total = df.groupby('Education_Sector')['Grand_Total'].sum().sort_values(ascending=False)
print("\nüéì Total Students by Education Sector:\n", sector_total)

# üìà Average Students per School
avg_enroll = df['Grand_Total'].mean()
print(f"\nüìà Average Students per School: {avg_enroll:.2f}")

# -------------------------------------------
# 5Ô∏è‚É£ Gender-based Analysis
# -------------------------------------------
# Try to extract male/female data if present
male_cols = [c for c in df.columns if 'Males' in c]
female_cols = [c for c in df.columns if 'Females' in c]

df['Total_Males'] = df[male_cols].sum(axis=1)
df['Total_Females'] = df[female_cols].sum(axis=1)

gender_total = pd.Series({
    'Males': df['Total_Males'].sum(),
    'Females': df['Total_Females'].sum()
})

print("\nüë¶üëß Total Students by Gender:\n", gender_total)

# -------------------------------------------
# 6Ô∏è‚É£ Visualization
# -------------------------------------------
plt.style.use('seaborn-v0_8')

# Bar chart: Students by School Type
plt.figure(figsize=(8,5))
school_type_total.plot(kind='bar', color='teal')
plt.title('Total Students by School Type')
plt.ylabel('Total Students')
plt.xlabel('School Type')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

# Pie chart: Education Sector Distribution
plt.figure(figsize=(6,6))
sector_total.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Education Sector Distribution')
plt.ylabel('')
plt.show()

# Scatter: School_No vs Grand_Total
plt.figure(figsize=(10,5))
sns.scatterplot(x='School_No', y='Grand_Total', data=df, hue='Education_Sector')
plt.title('Enrollment by School Number')
plt.xlabel('School Number')
plt.ylabel('Total Students')
plt.grid(True)
plt.show()

# Pie chart: Gender Distribution
plt.figure(figsize=(5,5))
gender_total.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue','lightpink'])
plt.title('Gender Distribution (Overall)')
plt.ylabel('')
plt.show()

# -------------------------------------------
# 7Ô∏è‚É£ Summary Insights
# -------------------------------------------
print("\nüìä Summary Insights:")
print(f"- Highest enrolled School Type: {school_type_total.idxmax()} ({school_type_total.max():.0f} students)")
print(f"- Dominant Education Sector: {sector_total.idxmax()} ({sector_total.max():.0f} students)")
print(f"- Gender Ratio (M:F): {gender_total['Males']:.0f}:{gender_total['Females']:.0f}")
```
